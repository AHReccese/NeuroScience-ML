### A2. Perceptron Algorithm

In this part, you should implement the perceptron algorithm from the scratch. First, add one dimension with <code>constant 1</code> to each data point (i.e. $x_0 = 1$). What is the purpose of this? <br>
Perform the algorithm for <code>50000</code> iterations, and at each <code>500</code> iterations calculate and save the error ($\frac{1}{n}\sum_{i=1}^{n} \boldsymbol{1}_{prediction_{i}\neq y_i}$) on the test data, and in the end plot the error against the number of iterations. After the end of the training process, report the final error on the test data and the final weights <code><bold>w</bold></code>.

```python
## Implement the algorithm here
import numpy as np
import random

def perceptron(X_tr, Y_tr, X_te, Y_te, max_iter=50000):
    
    w = np.array([0 for _ in range(len(X_tr[0]))]) 
    loss_history = list()

    for t in range(1,max_iter + 1):
        
        # Searching randomly
        indexes = random.sample(range(0,len(X_tr)),len(X_tr))
        # if you dont want to use randomness : replace it with below 
        # indexes = [i for i in range(len(X_tr))]
        
        # finding miss classify
        for i in indexes:
            predict = np.dot(X_tr[i],w)
            if( predict * Y_tr[i] <= 0 ):
                w = w + np.multiply(Y_tr[i],X_tr[i])
                # normalization: w = w / np.linalg.norm(w) 
                break
                
        if(t%500 == 0):
            error = 0
            for i in range(len(X_te)):
                predict = np.sign(np.dot(X_te[i],w))
                if(predict != Y_te[i]):
                    error = error + 1
            error = error/len(X_te)
            loss_history.append(error)
              
    return w, loss_history
```

```python
## train of training samples
w, loss_history = perceptron(x_train, y_train, x_test, y_test)

print("final W is: ")
print(w)
print("final Loss is: ")
print(loss_history[-1])
```

```python
## Plot test error over every 500 iterations
import matplotlib.pyplot as plt 
  
# line 1 points 
iteration = [i*500 for i in range(1,len(loss_history)+1)] 

# plotting the line 1 points  
plt.plot(iteration, loss_history, label = "loss History") 

# naming the x axis 
plt.xlabel('x - axis') 
# naming the y axis 
plt.ylabel('y - axis') 
# giving a title to my graph 
plt.title('Loss over iteration') 
  
# show a legend on the plot 
plt.legend(['Loss']) 
  
# function to show the plot 
plt.show() 
```

### B1. Load Data
The first cloumn is the label, and the remaining <code>22</code> columns are features of data points. Split the data into three sets: the first <code>70%</code> for training, the next <code>20%</code> for verification, and the remaining <code>10%</code> for testing. The validation set is for choosing the best model among all models based on the error ($\frac{1}{n}\sum_{i=1}^{n} \boldsymbol{1}_{prediction_{i}\neq y_i}$). The test set is for estimating the true error of the selected model.<br>
Import data with pandas library.

```python
import pandas as pd

data = pd.read_csv("Datasets\mushrooms.csv") 
length = len(data)

# considering 70 percent of total data to be our training Set.
# 20 percent to be our verification Set
# last 10 percent to be our test Set
_70percent = int(0.7*length)
_90percent = int(0.9*length)
 
trainSet = data[0:_70percent]
verificationSet = data[_70percent:_90percent]
testSet = data[_90percent:]

cols = data.columns
x_train = trainSet[cols[1:]].to_numpy() # Index(['feature 0','feature 1', 'feature 2', 'feature 3', 'feature 4'], dtype='object')
y_train = trainSet[cols[0]].to_numpy() # Label

x_val = verificationSet[cols[1:]].to_numpy() # Index(['feature 0','feature 1', 'feature 2', 'feature 3', 'feature 4'], dtype='object')
y_val = verificationSet[cols[0]].to_numpy() # Label

x_test = testSet[cols[1:]].to_numpy() # Index(['feature 0','feature 1', 'feature 2', 'feature 3', 'feature 4'], dtype='object')
y_test = testSet[cols[0]].to_numpy() # Label

data.head()
```

### B2. Train Desicion Tree
Set the <code>maximum depth</code> of the tree to $\{4, 6, 8, 10, 12, 14, 16, 18, 20\}$. For each maximum depth, train a classifier on the training data and report the resulting loss on the validation set. Plot the loss against the maximum depth. What is the best maximum depth?<br>
Finally, for the best maximum depth, report the loss on the test set.

```python
## Train Desicion Tree for each depth here
# Load libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier

def calculateErrorOnSet(classifier,X,Y):
    predict = classifier.predict(X)
    error = 0
    for i in range(len(predict)):
        if(predict[i] != Y[i]):
            error = error + 1
    error = error / len(predict)
    return error
        
depth_tree = [4, 6, 8, 10, 12, 14, 16, 18, 20]
clfs = []
validSetErrors = []

numberOfIteration = 1000
for _ in range(numberOfIteration):
    for i in range(len(depth_tree)):
        # Create Decision Tree classifer object
        # in order to avoid randomNess among classifiers -> setting random_state to 1(or any other integer)
        clf = DecisionTreeClassifier(max_depth = depth_tree[i]) #,random_state = 1)
        # Train Decision Tree Classifer
        clf = clf.fit(x_train,y_train)
        validSetErrors.append(calculateErrorOnSet(clf,x_val,y_val))
        clfs.append(clf)

avgError = [0]*9
for j in range(numberOfIteration):
    for i in range(9):
        avgError[i] = avgError[i] + validSetErrors[9*j+i]
        
avgError = [loss/numberOfIteration for loss in avgError]
print("So the best maximum depth for this Run is: 6 ")
print("For the best maximum depth, the loss on the test set is: ")
print(avgError[1])
```

```python
from sklearn import tree

for i in range(9):
    fig = plt.figure(figsize=(10,2))
    tree.plot_tree(clfs[i], 
                       feature_names=cols[1:],  
                       class_names=['poisonous','edible'],
                       filled=True)
```
